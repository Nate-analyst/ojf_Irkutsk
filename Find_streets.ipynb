```python
import requests
import pandas as pd
```

```python
url = 'https://dom.gosuslugi.ru/nsi/api/rest/services/nsi/fias/v4/streets?cityCode=8eeed222-72e7-47c3-ab3a-9a553c31cf72&itemsPerPage=1000&page=2&regionCode=6466c988-7ce3-45e5-8b97-90ae16cb1249'

response = requests.get(url)
result = response.json()
```

```python
result
```

```python
l =[]
for i in range(len(result)):
    street = result[i]['formalName']
    preffix = result[i]['shortName']

    l.append({
        'street':street,
        'preffix':preffix
    })

df = pd.DataFrame(l)
df
```

```python
df["street"] = df["preffix"] + ". " + df["street"]
df.drop(columns=["preffix"], inplace=True)
```

```python
df.drop_duplicates()
```

```python
df.to_excel('streets.xlsx')
```
